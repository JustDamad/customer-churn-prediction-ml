### MACHINE LEARNING

## FEATURE ENGINEERING
### BUSINESS OBJECTIVE:

Prepare the dataset for machine learning by converting the target variable and removing unnecessary features.

```python
## First, we have to Convert the Target Variable
## Most Machine Learning require numerical targets
## That means we have to Convert Churn to Numerical Values

df["Churn"] = df["Churn"].map({
    "No" : 0,
    "Yes" : 1
})

df["Churn"].value_counts()
```

```python
## Let's Remove Customer ID (This is because it does not contain any predictive information)

df = df.drop("customerID", axis = 1)

df.head()
```

```python
## Let's Verify Data Types

df.dtypes
```

```python
## Now we can go ahead to Separate Features and Target

x = df.drop("Churn", axis = 1)

y = df["Churn"]

print("Features shape :", x.shape)
print("Target shape :", y.shape)
```

```python
## Since Most Machine cannot work directly with text categories, we'll have to convert them into numerical columns.

x =pd.get_dummies(
    x,
    drop_first = True
)

x.head()
```

### The Reason for drop_first = True:

One hot encoding transform categorical variables into binary indicator variables. Using **drop_first = True** avoids multicollinearity by removing one redundant category from each encoded feature while preserving all neccessary information.

```python
## It's time for the TRAIN-TEST SPLIT

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size = 0.20,
    random_state = 42,
    stratify = y
)

print("Training set:", x_train.shape)
print("Training set:", x_test.shape)
```

```python
## Now, We move to Feature Scaling

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_train_scaled = x_train.copy()
x_test_scaled = x_test.copy()

numeric_columns = [
    "tenure",
    "MonthlyCharges",
    "TotalCharges"
]

x_train_scaled[numeric_columns] = scaler.fit_transform(
    x_train[numeric_columns]
)

x_test_scaled[numeric_columns] = scaler.transform(
    x_test[numeric_columns]
)
```

## MODELING STAGE

```python
## LOGISTIC REGRESSION (BASELINE)
## Let's import the Model

from sklearn.linear_model import LogisticRegression
```

```python
## Time to Train the Model

log_model = LogisticRegression(
    random_state = 42,
    max_iter = 1000
)

log_model.fit(
    x_train_scaled,
    y_train
)
```

```python
## Now, we can make predictions

y_pred = log_model.predict(x_test_scaled)

y_pred_prob = log_model.predict_proba(x_test_scaled)[:, 1]
```

```python
## We'll Evaluate the Model

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)

print("Accuracy :", round(accuracy_score(y_test, y_pred), 3))
print("Precision:", round(precision_score(y_test, y_pred), 3))
print("Recall   :", round(recall_score(y_test, y_pred), 3))
print("F1 Score :", round(f1_score(y_test, y_pred), 3))
```

```python
## Classification Report

from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
## Confusion Matrix

from sklearn.metrics import ConfusionMatrixDisplay

ConfusionMatrixDisplay.from_estimator(
    log_model,
    x_test_scaled,
    y_test,
    cmap = "Blues"
)

plt.title("Confusion Matrix - Logistic Regression")
plt.show()
```

## BUSINESS INTERPRETATION

A telecom company usually values **Recall** because missing customers who are about to leave can be more costly than offering an unnecessary retention incentive.

This Model has:

- Good overall accuracy (80.5%)
- Good precision (65%)
- Moderate recall (57.2%)

This means the Model is conservative. It avoids too many false alarms but misses some customers who eventually churn.

## Logistic Regression Model Evaluation

- The Logistic Regression Model achieved an accuracy of **80.5%**, indicating strong overall predictive performance.
  
- The Model recorded a precision of **65.0**, meaning that approximately two out of every three customers predicted to churn actually did so.

- It's recall of **57.2%** shows that the model successfully identified more than half of the customers who eventually churned. While this provides solid baseline, it also indicates that a significant proportion of churning customers remain undetected.

- The confusion matrix reveals that the model correctly classified **918 loyal customers**  and **214 churning customers**, while incorrectly classifying **115 loyal customers as churners** and missing **160 actual churn case**.

- From a business perspective, false negatives represent missed retention opportunities and potential revenue loss. Therefore, improving recall should be a priority in subsequent models.

- Over all, Logistic Regression provides a strong baseline model due to it's simplicity, interpretability, and competitive predictive performance.
